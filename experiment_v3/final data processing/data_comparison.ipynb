```python
import pandas as pd

# Load the dataset
df = pd.read_csv("dse_2008_2022.csv")

# Convert the date column to datetime
df["date"] = pd.to_datetime(df["date"])

# Filter CITYBANK records between Jul 12, 2021 and Dec 29, 2022 (inclusive)
citybank_df = df[
    (df["trading_code"] == "CITYBANK") &
    (df["date"] >= "2021-07-12") &
    (df["date"] <= "2022-12-29")
]

# Count the rows
print(f"Total CITYBANK rows between 2021-07-12 and 2022-12-29: {len(citybank_df)}")
```

```python
import json
from datetime import datetime, timezone

# 1) Path to your JSON file
DATA_FILE = "citybank_counts.json"

# 2) Set your date range (inclusive)
START_DATE = "2021-07-12"
END_DATE = "2022-12-29"

# --- processing ---
with open(DATA_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

def to_date(ts_ms):
    return datetime.fromtimestamp(ts_ms / 1000, tz=timezone.utc).date()

start = datetime.strptime(START_DATE, "%Y-%m-%d").date()
end = datetime.strptime(END_DATE, "%Y-%m-%d").date()

in_range = sorted(
    [(to_date(row[0]), row) for row in data if start <= to_date(row[0]) <= end],
    key=lambda x: x[0]
)

zero_days = [d for d, row in in_range if row[1] == 0 and row[2] == 0 and row[3] == 0]

print(f"Total rows in source data        : {len(data)}")
print(f"Rows between {START_DATE} and {END_DATE} : {len(in_range)}")
if in_range:
    print(f"First date in range : {in_range[0][0]}")
    print(f"Last date in range  : {in_range[-1][0]}")
if zero_days:
    print(f"No-trade / halted days ({len(zero_days)}): {zero_days}")
```
